```python
df["Main_Category"] = (
    df["product_category_tree"]
    .str.extract(r'\["([^>]*)')
)
```

```python
df.isnull().sum()
```

```python
df.drop(
    columns=['product_rating',
             'overall_rating',
             'image'],
    inplace=True
)
```

```python
df["description"]=df["description"].fillna("Not_Available")
```

```python
df['product_specifications'] = (
    df['product_specifications']
    .fillna('Not Available')
)
```

```python
df.isnull().sum()
```

```python
df = df.dropna(
    subset=['retail_price_₹',
            'discounted_price_₹',
           'Selling Price ₹',
           'Discount %']
)
```

```python
# price Segement 
```

```python
def price_segment(price):
    if price < 500:
        return 'Budget'
    elif price < 2000:
        return 'Mid-Range'
    else:
        return 'Premium'
```

```python
df['Price_Segment'] = (
    df['discounted_price_₹']
    .apply(price_segment)
)
```

```python
df.head(4)
```

```python
df.info()
```

```python
df['Discount_Amount'] = (
    df['retail_price_₹']
    - df['discounted_price_₹']
)
```

```python
df.head(3)
```

```python
df['Discount_Percentage'] = (
    df['Discount_Amount']
    / df['retail_price_₹']
) * 100
```
